```python
# Cell 1 — Create a Spark DataFrame
data = [
    ("Alice", "Data Engineer", 92),
    ("Bob", "ML Engineer", 88),
    ("Carol", "Data Scientist", 95),
    ("Dave", "Analytics", 78),
    ("Eve", "ML Engineer", 91)
]
columns = ["name", "role", "score"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
# Cell 2 — Basic DataFrame operations
print(f"Total rows: {df.count()}")
print(f"Schema:")
df.printSchema()
```

```python
# Cell 3 — Filter (like SQL WHERE)
df.filter(df.score > 90).show()
```

```python
# Cell 4 — GroupBy (like SQL GROUP BY)
df.groupBy("role").avg("score").show()
```

```python
# Cell 5 — Select specific columns
df.select("name", "score").orderBy("score", 
           ascending=False).show()
```
